```python
#Question 3
#part a
total elements=5
total number of possible outcomes=5!=>120
For comparison you need two values at a time, so 2^x>=120
Resolving above equation x should be atleast 7 (2^7=128 and 2^6=64)
Therefore, a binary comparison for the sorting procedure will have at least 7 levels.
```

```python
#part b
#algorithm for maximum 7 comparison to sort 5 elements

1. compare first two pairs and swap (2 comparison)
2. compare larger elements of first two swaps (1 comparison) and make new array which includes current winner,
   current loser and other element of loser pair. Now we have 3 sorted elements.
3. insert last element in 3 sorted elements by binary search (2 comparison)
4. insert other element of winner pair in step 2 before its pair element by binary search. (2 comparison)

#Example
arr=[5,4,3,2,1] #a1,a2,a3,a4,a5

1. [4,5,2,3] (max 2 comparisons) #a1>a2 and a3>a4
2. [5,3,2] (max 1 comparison) #a1>a3
3. [5,3,2,1] (max 2 comparison) #binary search insertion
4. [5,4,3,2,1] (max 2 comparison) #binary search insertion in last 3 elements as we know a2<a1 (from step 1)

so array can be sorted with maximum of 7 comparisons using this algorithm
```

```python
####code of question 2
INF = 1000000
import sys

def min(x, y):
    if x < y:
        return x
    return y

def minCoins(coins, m, V): 
      
    # table[i] will be storing the minimum  
    # number of coins required for i value.  
    # So table[V] will have result 
    table = [0 for i in range(V + 1)] 
  
    # Base case (If given value V is 0) 
    table[0] = 0
  
    # Initialize all table values as Infinite 
    for i in range(1, V + 1): 
        table[i] = sys.maxsize 
  
    # Compute minimum coins required  
    # for all values from 1 to V 
    for i in range(1, V + 1): 
          
        # Go through all coins smaller than i 
        for j in range(m): 
            if (coins[j] <= i): 
                sub_res = table[i - coins[j]] 
                if (sub_res != sys.maxsize and 
                    sub_res + 1 < table[i]): 
                    table[i] = sub_res + 1
    return table[V]

def findMin(d,V): 
      
    # All denominations of Indian Currency 
    deno = d
    n = len(deno) 
      
    # Initialize Result 
    ans = [] 
  
    # Traverse through all denomination 
    i = n - 1
    while(i >= 0): 
          
        # Find denominations 
        while (V >= deno[i]): 
            V -= deno[i] 
            ans.append(deno[i]) 
  
        i -= 1
    return len(ans)

if __name__ == '__main__':
    # array starting from 1, element at index 0 is fake
    d = [1,7,14,20,61]
    n=62
    for x in range(100000):     
        dyn=minCoins(d, len(d), n)
        gre=findMin(d,n)
        print(dyn,gre)
        if dyn!=gre:
            print("breakkkkk",n)
            break
        n+=1
```
